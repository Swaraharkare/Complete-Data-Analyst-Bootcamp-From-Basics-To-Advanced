```python
def transpose_3x3(matrix):
    """
    Takes a 3x3 matrix (nested list) and returns its transpose.
    The zip(*matrix) unpacks the rows and zips them into columns.
    """
    return [list(row) for row in zip(*matrix)]

# Define a sample 3x3 matrix
original_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Get the transposed matrix
transposed_matrix = transpose_3x3(original_matrix)

# Print the original matrix
print("--- Original Matrix ---")
for row in original_matrix:
    print(row)

# Print a blank line for separation
print()

# Print the transposed matrix
print("--- Transposed Matrix ---")
for row in transposed_matrix:
    print(row)
```
