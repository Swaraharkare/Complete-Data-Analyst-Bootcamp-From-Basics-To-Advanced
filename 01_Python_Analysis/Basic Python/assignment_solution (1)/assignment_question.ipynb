**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
n=5
fact = 1
for i in range(1, n + 1):
    fact *= i
print(fact) 
```

```python
import math

n = 5
result = math.factorial(n)
print(f"The factorial of {n} is {result}")  # Output: 120
```
